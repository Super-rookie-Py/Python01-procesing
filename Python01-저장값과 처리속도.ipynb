### 변환과정으로 새로운 리스트를 만들기 (리스트 내포)

#### [표현식 for i in 반복가능객체 if 조건문1for 항목2 in 반복객체 2 if 조건2]

```python
myList = [1, 2, 3, 4]
```

```python
y = [x for x in myList]
```

```python
# 리스트의 데이터값을 새로운 리스트형태로 저장하기.

z = [x for x in myList if x>2]
z
```

```python
[x for x in myList if x%2 == 0]
```

```python
[x for x in myList if x%2 == 1]
```

```python
[x**2 for x in myList]
```

#### 앞에서 사용한 출력식은 뒤의 값과 동일해야한다.

#### unit07 counter

```python
from collections import Counter
```

```python
p = "a man a plan a canal panama"
cntr = Counter(p.split())

print(p.split())
# 값의 빈도를 나타냄
print(cntr)

# 결과 값의 개수를 나타내는식
# 리스트안에 튜플값들이 저장되어있음.
cntr.most_common()
```

```python
# 다음 값들을 딕셔너리형태로 저장

cntrDict = dict(cntr.most_common())
cntrDict
```

```python
cntrDict['a']
```

#### 파일다루기

```python
f = open("test.txt", mode = "r")
```

```python
# 파일을 읽었으면 꼭 닫아줘야 다른 곳에서 사용권한이 생김

f.read()
```

```python
f.close()
```

```python
# 파일을 닫지않아도 with 함수에서 벗어나게됨

with open("test.txt", mode = "r") as f:
    print(f. read())
```

```python
f = open("test.txt", mode = "r")
lines = f.readlines()
for line in lines:
    print(line)
f.close()
```

```python
with open("test.txt", mode = "r") as f:
    lines = f.readlines()
    for line in lines:
        print(line)
```

#### 결론 파일을 열고 닫을 때는 with as 를 사용하는 것이 효과적이다
